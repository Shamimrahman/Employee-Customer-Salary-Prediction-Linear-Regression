```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('../input/random-salary-data-of-employes-age-wise/Salary_Data.csv')
```

```python
df.describe().round(2)
```

```python
plt.scatter(df['YearsExperience'],df['Salary'],color='red')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Employee Salary Prediction')
```

```python
x=df[['YearsExperience']]
y=df['Salary']
```

```python
xtest,xtrain,ytest,ytrain=train_test_split(x,y,test_size=.3)
```

```python
lg=LinearRegression()
```

```python
lg
```

```python
lg.fit(xtrain,ytrain)
```

```python
accuracy=lg.score(xtest,ytest) 
```

```python
print('The Accuracy is ', accuracy)
```

```python
#prediction

n=input('What is your Experience')
array=np.array(n)
array2=array.astype(np.float)

value=[[array2]]

predict=lg.predict(value).round(2)

result=np.array(predict)

result=result.item()

print("The Predicted Salary is" , result)
```
